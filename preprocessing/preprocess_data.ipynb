```python
ll_data_path = "../../autodan-itau/tests/results/lifelong/attack_logs.json"
wm_data_path = "../../autodan-itau/tests/results/warmup/attack_logs.json"
```

```python
import pandas as pd
import numpy as np

data_ll = pd.read_json(ll_data_path)
data_wm = pd.read_json(wm_data_path)
```

```python
data = pd.concat([data_wm, data_ll])
```

```python
data
```

```python
new_data = data[["malicious_request", "attack_prompt", "score"]].copy()
```

```python
new_data.sort_values(by="score", ascending=True, inplace=True)
```

```python
new_data.reset_index(drop=True, inplace=True)
```

```python
new_data.to_json("../data/data.json", orient="records", indent=4)
```
