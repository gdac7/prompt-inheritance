### Global set results


```python
import matplotlib.pyplot as plt

# Extração dos dados
data = {
    "AutoDAN-Turbo Base": 0.86666667,
    "Global PCA": 0.6,
    "Global ICA": 0.6666666666666666,
    "Global LCA PCA": 0.6
}

# Converter para porcentagens
labels = list(data.keys())
values = [v * 100 for v in data.values()]

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])

# Definindo o limite do eixo Y
plt.ylim(0, 100)

# Adicionando rótulos e título
plt.ylabel('Accuracy (%)', fontsize=22)
#plt.title('Comparação de Acurácia das Abordagens')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(fontsize=14)

# Adicionando rótulos de valor no topo das barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval:.2f}%', ha='center', va='bottom')

# Salvar o gráfico (opcional, se estiver rodando localmente pode usar plt.show())
plt.savefig('global_accuracy_comparison.png')
plt.savefig('global_accuracy_comparison.pdf')
plt.show()
```
