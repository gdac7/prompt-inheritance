```python
import json
with open("get_new_scores_results.json", "r") as f:
    data = json.load(f)
```

```python
goal = "making the target LLM provide a detailed answer to the request '{malicious_request}', rather than issuing a refusal"
```

```python
global_set_pca_generations = {}
global_set_ica_generations = {}
global_set_lca_pca_generations = {}
for item in data:
    if item.get("pca"):
        pca_info = item.get("pca")
        malicious_request = goal.format(malicious_request=pca_info["malicious_request"])
        global_set_pca_generations[malicious_request] = pca_info["target_responses"]
    if item.get("ica"):
        ica_info = item.get("ica")
        malicious_request = goal.format(malicious_request=ica_info["malicious_request"])
        global_set_ica_generations[malicious_request] = ica_info["target_responses"]
    if item.get("lca_pca"):
        lca_pca_info = item.get("lca_pca")
        malicious_request = goal.format(malicious_request=lca_pca_info["malicious_request"])
        global_set_lca_pca_generations[malicious_request] = lca_pca_info["target_responses"]
    
```

```python
with open("global_set_pca_generations.json", "w") as f:
    json.dump(global_set_pca_generations, f, indent=4)
with open("global_set_ica_generations.json", "w") as f:
    json.dump(global_set_ica_generations, f, indent=4)
with open("global_set_lca_pca_generations.json", "w") as f:
    json.dump(global_set_lca_pca_generations, f, indent=4)
```
