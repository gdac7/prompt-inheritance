```python
import json
with open("attack_logs_100requests.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
import pandas as pd

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
new_data = [
    {
        "malicious_request": row.malicious_request,
        "attack_prompt": row.attack_prompt,
        "score": row.score,
    }
    for row in df.itertuples(index=False)
]
```

```python
with open("../data/data.json", "w", encoding="utf-8") as f:
    json.dump(new_data, f, indent=4)
```
