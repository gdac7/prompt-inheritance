```python
import json
import pandas as pd

# 1. Carregar os arquivos
with open('autodan_metrics.json', 'r') as f:
    autodan_data = json.load(f)
with open('scorer_performance_metrics.json', 'r') as f:
    scorer_data = json.load(f)

# 2. Converter as m√©tricas detalhadas para DataFrames
df_autodan = pd.DataFrame(autodan_data['detailed_metrics'])
df_scorer = pd.DataFrame(scorer_data['detailed_metrics'])

# 3. Concatenar e Ordenar
# O pandas vai alinhar as colunas automaticamente.
df_merged = pd.concat([df_autodan, df_scorer], ignore_index=True)
df_merged['timestamp'] = pd.to_datetime(df_merged['timestamp'])
df_merged = df_merged.sort_values('timestamp').reset_index(drop=True)

# 4. (Opcional) Salvar de volta em JSON ou CSV
# Se quiser manter a estrutura original do JSON do AutoDAN:
autodan_data['detailed_metrics'] = df_merged.to_dict(orient='records')

# Nota: Os objetos timestamp precisam voltar a ser string para salvar em JSON
for item in autodan_data['detailed_metrics']:
    item['timestamp'] = item['timestamp'].isoformat()

with open('autodan_metrics_completo.json', 'w') as f:
    json.dump(autodan_data, f, indent=4)
```

```python
import json
with open("autodan_generations.json", "r") as f:
    adn = json.load(f)
with open("pca_generations.json", "r") as f:
    pca = json.load(f)
```

```python
len(adn)
```

```python
len(pca)
```

```python
new_autodan_generations = {}
for key, value in adn.items():
    if pca.get(key) is not None:
        new_autodan_generations[key] = value
```

```python
with open("ica_generations.json", "r") as f:
    ica = json.load(f)
with open("swpca_generations.json", "r") as f:
    swpca = json.load(f)
with open("gw_generations.json", "r") as f:
    gw = json.load(f)
new_pca_generations = {}
new_ica_generations = {}
new_swpca_generations = {}
new_gw_generations = {}
for key, value in new_autodan_generations.items():
    if pca.get(key) is not None:
        new_pca_generations[key] = value
    if ica.get(key) is not None:
        new_ica_generations[key] = value
    if swpca.get(key) is not None:
        new_swpca_generations[key] = value
    if gw.get(key) is not None:
        new_gw_generations[key] = value
    
```

```python
with open("new_generations/new_autodan_generations.json", "w") as f:
   json.dump(new_autodan_generations, f, indent=4)
with open("new_generations/new_pca_generations.json", "w") as f:
    json.dump(new_pca_generations, f, indent=4)
with open("new_generations/new_ica_generations.json", "w") as f:
    json.dump(new_ica_generations, f, indent=4)
with open("new_generations/new_swpca_generations.json", "w") as f:
    json.dump(new_swpca_generations, f, indent=4)
with open("new_generations/new_gw_generations.json", "w") as f:
    json.dump(new_gw_generations, f, indent=4)
```
